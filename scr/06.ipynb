### 여러 조건을 충족하는 행 추출하기

```python
# 1반이면서 수학 점수가 50 점 이상인 경우 :   & , and 연산자 사용가능
exam.query('nclass == 1  &   math >= 50')
```

```python
# 2반이면서 영어 점수가 80 점 이상인 경우
exam.query('nclass == 2 & english >= 80')
```

### 여러 조건 중 하나 이상 충족하는 행 추출하기

## 06-6 집단별로 요약하기

### 집단별로 요약하기

#### 전체 요약 통계량 구하기


```python
# math 평균 구하기
exam.agg(mean_math = ('math', 'mean'))
# exam['math'].mean() 와 같은 의미
```

#### 집단별 요약 통계량 구하기



```python
# nclass별로 분리하기
# math 평균 구하기
```

```python
# nclass 별로 math 성적의 평균을 구한다. (nclass값이 인덱스 값으로 사용된다.)
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))
```

------------------------------------------------------------------------

#### (알아 두면 좋아요) 변수를 인덱스로 바꾸지 않기

```python
# exam.groupby() 함수에 as_index=False 를 설정하면, nclass값을 인덱스로 사용하지 않고 원래 인덱스를 
# 유지 한다는 의미를 가진다.
exam.groupby('nclass', as_index = False) \
    .agg(mean_math = ('math', 'mean'))
```

------------------------------------------------------------------------

#### 여러 요약 통계량 한 번에 구하기



```python
# nclass별로 분리
# 수학 점수 평균
# 수학 점수 합계
# 수학 점수 중앙값
# 빈도(학생 수)
```

```python
exam.groupby('nclass') \
    .agg(mean_math   = ('math', 'mean'),     
         sum_math    = ('math', 'sum'),
         median_math = ('math', 'median'),          
         n           = ('nclass', 'count'))
```

#### (알아 두면 좋아요) `value_counts()`로 집단별 빈도 간단하게 구하기

```python
# drv(구동방식)별로 나눠서 count함수로 빈도수를 구한다.
mpg.groupby('drv') \
   .agg(n = ('drv', 'count'))
```

```python
# drv(구동방식) 변수에 있는 데이터의 빈도수를 value_counts()함수로 구한다.
# value_counts()함수는 빈도수를 기준으로 내림차순 정렬한다.
mpg['drv'].value_counts()
```

```python
# value_count()함수로 구해진 결과는 시리즈(series) 자료구조 이므로 query()함수를 적용할 수 없다. 
mpg['drv'].value_counts().query('n > 100')       # 오류 발생
```

```python
# drv 빈도 구하기
# 데이터 프레임으로 바꾸기, 변수명 n으로 바꾸기
# axis 이름에 drv 지정
# n이 100을 초과한 경우 추출
```

```python
# 빈도(n)이 100보다 큰 데이터만 검색
# df.value_counts()함수의 출력 결과에 query()함수를 적용하기 위해서는 데이터프레임으로 변환해야 한다.
mpg['drv'].value_counts() \
          .to_frame('n') \
          .rename_axis('drv') \
          .query('n > 100')
```

### `pandas` 함수 조합하기

```python
# suv 추출
# 합산 연비 변수 만들기 (고속도로 연비 + 도시연비)
# 제조 회사별로 분리
# 합산 연비 평균 구하기
# 내림차순 정렬
# 1~5위까지 출력
```

```python
mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head()
```

## 06-7 데이터 합치기

### 가로로 합치기

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 4, 5],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1  # test1 출력
```

```python
test2  # test2 출력
```

```python
# id 기준으로 합쳐서 total에 할당
# merge() : 2개의 데이터프레임을 가로로 합칠때 사용한다.
# how = 'left' : 오른쪽에 데이터프레임(test2)을 왼쪽 데이터프레임(test1)에 결합한다.
# on = 'id'    : 데이터를 합칠때 'id' 변수를 기준으로 데이터프레임을 합친다.
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

### 세로로 합치기

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})

# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python
group_a  # group_a 출력
```

```python
group_b  # group_b 출력
```

```python
# 데이터 합쳐서 group_all에 할당
# concat() : 2개의 데이터프레임을 세로로 합칠때 사용한다.
group_all = pd.concat([group_a, group_b])
group_all
```

### 정리하기

```python
## 1. 조건에 맞는 데이터만 추출하기
exam.query('english <= 80')

# 여러 조건 동시 충족
exam.query('nclass == 1 & math >= 50')

# 여러 조건 중 하나 이상 충족
exam.query('math >= 90 | english >= 90')
exam.query('nclass in [1, 3, 5]')

## 2. 필요한 변수만 추출하기
exam['math']                                 # 한 변수 추출
exam[['nclass', 'math', 'english']]          # 여러 변수 추출
exam.drop(columns = 'math')                  # 변수 제거
exam.drop(columns = ['math', 'english'])     # 여러 변수 제거

## 3. pandas 명령어 조합하기
exam.query('math >= 50')[['id', 'math']].head()

## 4. 순서대로 정렬하기
exam.sort_values('math')                     # 오름차순 정렬
exam.sort_values('math', ascending = False)  # 내림차순 정렬

# 여러 변수 기준 정렬
exam.sort_values(['nclass', 'math'], ascending = [True, False])


## 5. 파생변수 추가하기
exam.assign(total = exam['math'] + exam['english'] + exam['science'])

# 여러 파생변수 한 번에 추가하기
exam.assign(total = exam['math'] + exam['english'] + exam['science'],
            mean = (exam['math'] + exam['english'] + exam['science']) / 3)

# assign()에 np.where() 적용하기
exam.assign(test = np.where(exam['science'] >= 60, 'pass', 'fall'))

# 추가한 변수를 pandas 코드에 바로 활용하기
exam.assign(total = exam['math'] + exam['english'] + exam['science']) \
    .sort_values('total') \
    .head()

## 6. 집단별로 요약하기
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))

# 각 집단별로 다시 집단 나누기
mpg.groupby(['manufacturer', 'drv']) \
   .agg(mean_cty = ('cty', 'mean'))

## 7. 데이터 합치기
pd.merge(test1, test2, how = 'left', on = 'id')  # 가로로 합치기
pd.concat([group_a, group_b])                    # 세로로 합치기
```
